## Introduction

In classical mechanics a particle may be described by its position and velocity, but in quantum mechanics the particle’s state of motion is described using a wave equation. This projects delves into the idea of wave equations, their application in the solution for a spinless point particle in a quadratic potential well, and using numerical integration to calculate the expectation values for the quantum harmonic oscillator (QHO). 

In addition to there being an equation for a particles state of motion there is also an equation that models how the particle’s state of motion changes with time, this is called an equation of motion. In classical mechanics the equation of motion is $\vec{F}=m\vec{a}$, which given an initial position and velocity would allow the position and velocity to be calculated at any other time. In quantum mechanics this idea remains the same but instead of Newton’s second law as the equation of motion there is the time-dependent Schrödinger Equation, which allows the calculation of the wave function at any other time. In this project we will not use the time-dependent form of this equation, but will instead only refer to the time-independent form. 

As is the case with any law of physics, Schrödinger equation cannot be derived [1], it is instead a relation that has been observed to be the case. That being said it can be motivated by the fact most laws of physics can be expressed as differential equations which are equations that involve a combination of a variable and some of its derivatives. Due to the complexity of the solution for the QHO I will begin by introducing the simplified case of a particle in a rigid box to give a foundation for wave functions, I will then move onto more complicated cases until I get to the QHO. 

The wave function for a particle in a rigid box has the form

\begin{equation*}
\psi (x) = A \sin (kx).  \tag{Equ. 1}
\end{equation*}

Differentiating this twice gives us the equations

\begin{equation*}
\frac{d \psi}{dx} = kA \cos (kx), \tag{Equ. 2}
\end{equation*}

and 

\begin{equation*}
\frac{d^2 \psi}{dx^2} = -k^2 A \sin (kx). \tag{Equ. 3}
\end{equation*}

We can easily see that the equations Equ. 1 and Equ. 3 are proportional, and as such we can rewrite equation Equ. 3 as

\begin{equation*}
\frac{d^2 \psi}{dx^2} = -k^2 \psi \tag{Equ. 4}
\end{equation*}

Since $k$ is related to the particle’s momentum by $p = \hbar k$ and a particle's kinetic energy is related to the momentum based on $K = \frac{p^2}{2m}$, we can rewrite $k^2$ as

\begin{equation*}
k^2 = \frac{2mK}{\hbar^2}. \tag{Equ. 5}
\end{equation*}

The final equation needed to give us the differential equation is the relationship of kinetic energy with potential energy (U) and total energy (E)

\begin{equation*}
K \ = \ E \ - \ U(X), \tag{Equ. 6}
\end{equation*}

which is most commonly written as total energy being equal to the sum of the kinetic and the potential. Plugging our $K$-value into our $k$ equation, and then plugging our $k$ equation into Equ. 4 we find the second order differential equation

\begin{equation*}
\frac{d^2 \psi}{dx^2} = \frac{2m}{\hbar^2} \big [ U(x) \ - \ E \big ] \ \psi (x). \tag{Equ. 7}
\end{equation*}

For brevity, we will no longer write derivatives as fraction but instead we will use prime notation

\begin{equation*}
\frac{d \psi}{dx} = \psi’.
\end{equation*}

An important aspect of this differential equation is the relationship between $U(x)$ and $E$. Equation 7 shows that the sign of $\psi''(x)$ is dependant on whether the potential energy is greater than or less than the energy of the particle. The sign of $\psi''(x)$ will give us the concavity of $\psi(x)$ which will help explain why a quantum particle behaves different than a classical particle in its probability distribution.

In studying a quantum harmonic oscillator there are many situations that can be looked at, potentially the simplest is that of a particle with one dimensional motion in an infinite square well. This problem involves a particle existing in a confined region such as a rigid box where it can only move side to side. Classically speaking this case can be broken up into two regions. If we define the box as length $a$ and extending from position 0 to $a$ then the first region would be from 0 to $a$. This is the domain in which the particle can move and will have the potential energy based upon the region. Since potential energy is relative we can consider it zero. The second region has two separate sections which act the same, below 0 and above $a$, where the particle would need to have an infinite amount of potential energy to exists there. 
	
Moving to a more complicated system we can look at a finite square well which is simply the infinite square well except there exists a finite amount of energy needed to escape the energy well. We can further complicating the situation to have a more realistic case where the particle’s energy changes continuously at the bounds, this is often referred to as the rounded well. The important difference of the last two situations from the infinite square well is the fact there can exist an $E$-value which can be greater than the $U$-value. Where $U \ - \ E$ is positive is often referred to as the _classically forbidden region_ and where $U \ - \ E$ is positive is referred to as the _classically allowed region_. These are named in this way because classical particles cannot leave potential wells requiring more energy than they have to give.

A quantum particle has two very important differences from the classical particle for the finite well cases. When our quantum particle is within the classically defined region $U \ - \ E$ is positive, meaning
	
\begin{equation*}
\psi ’’ (x) = (positive \ function) \  \psi (x), \tag{Equ. 8}
\end{equation*}

or the wave function curves away from the x-axis. If $U \ - \ E$ is negative

\begin{equation*}
\psi ’’ (x) = (negative \ function) \ \psi (x), \tag{Equ. 9}
\end{equation*}

or the wave function curves towards the x-axis. Without proof the general solution is

\begin{equation*}
\psi (x) = A e^{\alpha \ x} + B e^{-\alpha \ x} \tag{Equ. 10}
\end{equation*}

which means that the wave function values decay exponentially in the classically forbidden regions. This is a huge difference between the classical and quantum particle: the fact a quantum particle has a probability of existing outside a well it does not have enough energy to leave. This is often referred to as quantum tunneling, and is the reason the figures below do not have a limited non-zero domain, but instead have function values that approach zero as the the x-values go towards positive and negative infinity. 

A further complicated case is the simple harmonic oscillator (SHO), which is the model given by an object oscillating about a point of stable equilibrium with a force given by the form of Hooke's law. This force can then be used to solve for a potential energy function and used to solve for the state of motion. The quantum harmonic oscillator is the solution given by the Schrödinger equation with a potential energy value given by
	
\begin{equation*}
U(x) = \frac{1}{2} kx^2. \tag{Equ. 11}
\end{equation*}

This is an expanded case of the curved potential well, a potential well described by a quadratic curve. In reality the energy well is not truly quadratic and is only accurate for small displacements. 

The solution to Schrödinger’s equation$^{\text{[1]}}$ **CHECK IF CITATION IS FOR 1 OR 2** is beyond the scope of this paper and as such will not be shown here. The solution gives us a key difference between the classical SHO and the quantum SHO: the quantum SHO has discrete values for the allowed energies. Where $\omega_c = \sqrt{k/m}$ is the frequency, the energy levels are given by
	
\begin{equation*}
E_n = (n \ + \ \frac{1}{2}) \hbar \omega_c \ \ \ n = 0, \ 1, \ 2, \ .... \tag{Equ. 12}
\end{equation*}

Simplifying the solution to all units of one we find the _n_ th energy level of the one dimensional quantum SHO is given by

\begin{equation*}
\psi_n (x) \ = \ \frac{e^{-x^2/2}}{\sqrt{2^nn! \sqrt{\pi}}} H_n (x), \tag{Equ. 13}
\end{equation*}

Where $H_n (x)$ is the _n_ th Hermite polynomial

\begin{equation*}
H_{n+1} (x) = 2x \ H_n (x) - 2n \ H(n-1) (x). \tag{Equ. 14}
\end{equation*}

Here $H_0 (x) = 1$ and $H_1 (x) = 2x$ are the first two Hermite polynomials. Attached is a functioning program that calculates both $\psi_n (x)$ and $H_n (x)$ for given x and n-values. In the program there are also functions to calculate the uncertainty of the wave equation using numerical integration methods. The uncertainty is that found in the Heisenberg's uncertainty principle $\Delta x \Delta p = \hbar / 2$, and in given by 

\begin{equation*}
\Delta x = \sqrt{<x^2> - <x>^2}, \tag{Equ. 15}
\end{equation*}

where

\begin{equation*}
<x> \ = \ \int^{\infty}_{-\infty} x \ |\psi_n(x)|^2 \ dx. \tag{Equ. 16}
\end{equation*}

and

\begin{equation*}
<x^2> \ = \ \int^{\infty}_{-\infty} x^2 \ |\psi_n(x)|^2 \ dx \tag{Equ.17}
\end{equation*}

are the expectation values. Expectation values are the values of a function $f(x)$ that we would expect to find after many samples given the probability distribution of $f(x)$ as $p(x)$. The general form of the equation is

\begin{equation*}
<f(x)> \ = \ \int^{\infty}_{-\infty} f(x) \ p(x) \ dx. \tag{Equ. 18}
\end{equation*}

In our case we are looking at a wave function which has a probability density given by the absolute value of the wave function squared. As we shall see the QHO has a value of zero for $<x>$, meaning that the uncertainty in the position is given by

\begin{equation*}
\Delta x = \sqrt{<x^2>}, \tag{Equ. 19}
\end{equation*}

or the root-mean-square of the wave function.

It is impossible to numerically calculate the root-mean-squared integral (Equ. 15) in its base state. The reason for this is because the integral has a domain from negative infinity to positive infinity, and we are unable to perform an infinite number of calculations in a finite amount of time. A way around this is to use a change of variables in order to change the domain of integration. The method used in this program involves replacing $x$ with $\tan (x)$, dividing the whole integral by $\cos^2 (x)$, and changing the domain to $(\frac{-\pi}{2}, \ \frac{\pi}{2})$. There are issues with this transformation as seen in figure 5, where regions of the function have large amplitudes with small widths. Decreasing the number of sample points we use to calculate the area under the curve drastically decreases the accuracy of our approximation. Here we can see the importance of picking an ample number of sample points when we perform our integration as a result of the large peaks in the function. Having small sample points (or N-values) results in an unsatisfactory approximation of the function as seen in the N = 50, 25, and 10 cases, where the maximum value of the function is fairly significantly different from the N = 1000 case.

```python
import gaussxw
import math
import pylab as py
import numpy as np
import romberg as rm

%matplotlib inline
```

```python
def hermitePolynomial (nValue, xValue):
    """
    Description: Calculates the nth Hermite polynomial for the given n-value at the given x-value
    
    Parameters: nValue - the nth Hermite polynomial to be calculated
                xValue - the independant variable where the Hermite polynomial will be determined
    
    Return: The value of the nth Hermite polynomial at the given x-value
    """
    
    hermiteCur = 0
    hermiteNext = 1
  
    for i in range (nValue):
        hermiteCur, hermiteNext = hermiteNext, 2 * xValue * hermiteNext -  2 * i * hermiteCur
    
    return hermiteNext
    
```

```python
def quantumWaveFunc (nValue, xValue):
    """
    Description: Calculates the wavefunction of the nth energy level of the one-dimensional quantum harmonic oscillator
    
    Parameters: nValue - the energy level being calculated
                xValue - the independant variable where the wavefunction is being evaluated
    
    Return: The value of the wavefunction for the nth energy level evaluated at the given x-value
    """
    
    return math.exp (-xValue**2 / 2) / (math.sqrt (2**nValue * math.factorial (nValue) * math.sqrt (math.pi))) * hermitePolynomial (nValue, xValue)
```

```python
labelFontSize = 20
```

```python
xLabel = "X-value"
yLabel = r"$\psi (x)$"
title = "QHO Wave functions"
figureSize = (10, 8)

numberOfDataPoints = 100
nValues = [0, 1, 2, 3]
x1 = -4
x2 =  4

py.figure (figsize = figureSize)

#calculate data for QSHO first four eigen states
xValues = np.linspace (x1, x2, numberOfDataPoints, endpoint = True)
yValues = np.empty ([len (nValues), numberOfDataPoints])

for i in range (len (nValues)):
    for j in range (len (xValues)):
        yValues[i][j] = quantumWaveFunc (nValues[i], xValues[j])

#plot QSHO first four eigen states
ax = py.subplot (1, 1, 1)
ax.set_xlabel (xLabel, fontsize = labelFontSize)
ax.set_ylabel (yLabel, fontsize = labelFontSize)
ax.set_title (title, fontsize = labelFontSize)
ax.plot (xValues, yValues[0], 'k-', label = "n = 0")
ax.plot (xValues, yValues[1], 'g-', label = "n = 1")
ax.plot (xValues, yValues[2], 'b-', label = "n = 2")
ax.plot (xValues, yValues[3], 'y-', label = "n = 3")
py.legend()
```

**Figure 1**

This figure shows the first four energy levels of the quantum SHO. The general shape of each wave function bares a remarkable resemblance to the wave function of a particle in a rigid box which in turn resembles a classical wave on a string. The main difference between the above wave functions and a standing wave is the values of the wave function for a standing wave must be zero in the classically forbidden regions. As we can see in this figure the wave function degrades in amplitude as we look at x-values far enough from the y-axis, but oscillates within the classically allowed regions. This is a result of the wave functions being given by a sinusoidal equation multiplied by a decay factor. The decay factor is why the ends of the wave functions trickle off towards zero. This trickling effect is due to the $U (x)$-values being greater than the $E_n$-value for the particle at the given x-position.

Subplot (1) shows the wave function for the n = 5 energy level. Since the square of the absolute value of a wave function is the probability distribution for the wave function, subplot (2) shows this.
```python
numberOfDataPoints = 10000

nValues = [1, 2, 3, 10, 21, 35]
x1 = -10
x2 =  10
figureSize = (17, 13)
plotLabelXPos = 0.05
plotLabelYPos = 0.9
subPlotNumber = 0
plotLabelFontSize = 20
labelFontSize2 = 14

py.figure (figsize = figureSize)

xValues = np.linspace (x1, x2, numberOfDataPoints, endpoint = True)

#plot expectation value distribution
for nValue in nValues:
    
    subPlotNumber += 1
    xLabel = "X-value"
    yLabel = r"$x \ |\psi_{} (x)|^2$".format (nValue)
    title = "QHO <x> Integrand n={}".format (nValue)

    yValues = np.empty ([numberOfDataPoints])

    for j in range (len (xValues)):
        yValues[j] = xValues[j] * quantumWaveFunc (nValue, xValues[j])**2

    ax = py.subplot (2, 3, subPlotNumber)
    ax.text (plotLabelXPos, plotLabelYPos, "(" + str (subPlotNumber) + ")", fontsize=plotLabelFontSize, \
             transform=ax.transAxes)
    ax.set_xlabel (xLabel, fontsize = labelFontSize2)
    ax.set_ylabel (yLabel, fontsize = labelFontSize2)
    ax.set_title (title, fontsize = labelFontSize2)
    ax.plot (xValues, yValues, 'k-')
```

**Figure 4**

This figure shows the $<x>$-values for the energy levels n = 1, 2, 3, 10, 21, and 35. This helps to show that the expectation value of $x$ will always be zero regardless of the energy level. This can be easily verified by looking at equation 16. Since $|\psi (x)|^2$ will be positive and symmetric about the y-axis, and our domain of interest is symmetric around the y-axis, the integral will always be zero. As a result of this $\Delta x$ is based entirely upon $<x^2>$.

```python
def rootMeanFunc (nValue, xValue):
    """
    Description: returns the integrand of the square of the root-mean-squared for a 
                    quantum wave function integration
    
    Parameters: nValue - nth value of the wave function
                xValue - independant variable at which this function is being evaluated
    
    Return:the integrand of the square of the root-mean-squared for a quantum wave function
    """
    return xValue**2 * (abs (quantumWaveFunc (nValue, xValue)))**2

def rootMeanSquaredFunc (numSamplePoints, nValue, x1, x2):
    """
    Description: returns the root-mean-squared of a wave function integration
    
    Parameters: numSamplePoints - number of sample points to evaluate the error, 
                                    larger values determine more accurate approximations
                nValue - nth value of the wave function
                x1 - left hand value of domain 
                x2 - right hand value of domain
    
    Return: the root-mean-squared of a wave function integration
    """
    integral = 0
    
    x, w = gaussxw.gaussxw (numSamplePoints)

    xp = 0.5 * (x2 - x1) * x + 0.5 * (x2 + x1)
    wp = 0.5 * (x2 - x1) * w

    for i in range (numSamplePoints):
        integral += wp[i] * rootMeanFunc (nValue, math.tan (xp[i])) / (math.cos (xp[i]))**2

    integral = math.sqrt (integral)
    
    return integral
```

```python
numSamplePoints = 100
x1 = -math.pi / 2
x2 = math.pi / 2
nValue = 5

print ("Uncertainty for n =", nValue, "case:", rootMeanSquaredFunc (numSamplePoints, nValue, x1, x2))
```

This uncertainty is based upon equation 17 and 19 and was verified by Newman. 

```python
xLabel = "X-value"
yLabel = "Transformed Error Integrand"
title = "Quantum Harmonic Oscillator Transformed n=5"
nValue = 5
x1 = -math.pi / 2
x2 =  math.pi / 2


numberOfDataPoints = [1000, 50, 25, 10]
legends = ['k-', 'g-', 'b-', 'r-']
subPlotSize = (12, 10)
baseTitleStr = "Transformed Integrand for "

py.figure (figsize = subPlotSize)

for i in range (len (numberOfDataPoints)):
    
    ax = py.subplot (2, 2, i + 1)
    
    x, w = gaussxw.gaussxw (numberOfDataPoints[i])

    xp = 0.5 * (x2 - x1) * x + 0.5 * (x2 + x1)
    wp = 0.5 * (x2 - x1) * w

    yValues = []
    
    titleStr = baseTitleStr + "N={}".format (numberOfDataPoints[i])
    
    for xValue in xp:
        yValues.append (rootMeanFunc (nValue, math.tan (xValue)) / (math.cos (xValue))**2)
    
    ax.set_xlabel ("Z-value")
    ax.set_ylabel ("Integrand Value")
    ax.set_title (titleStr)
    ax.plot (xp, yValues, legends[i])
```

**Figure 5**

This figure shows the integrand for $<x^2>$ after a variable substitution. The substitution was necessary to allow for a numerical solution to the integral over an infinite domain.   
A rather large issue that can arise from integrating a function like this is the fact there are large peaks with small widths. Numerically calculating the integral of a function tends to involve looking at the breaking the function up into small geometric pieces that are easy to find the area of. The issue is that the accuracy of our calculation is dependant upon the number of geometric pieces we use to calculate the area of the function. A small number of pieces with steep slopes  result in the heights of the geometric pieces being misrepresented. The solution to this can be to increase the number of geometric pieces used but this increases the number of calculations needed to occur which increases the runtime of our program.


```python
def rombergQuantumWaveFunc (waveFunc, nValue, x1, x2, tolerance):
    """
    Description: calculates the integral value of the given wave function and the rootMeanSquared for the specified nValue
    
    Parameters: waveFunc - wavefunction being analysed
                nValue - energy level of wave function being checked
                x1 - left most bound
                x2 - right most bound
                tolerance - tolerance for the romberg integration method
    
    return: the integral value of the given wave function and the uncertainty of the specified energy level
    """
    integral = 0
    rootMeanSquared = 0
    errorSamplePoints = 100
    
    def waveFuncSetN (xValue):
        return waveFunc (nValue, xValue)
    
    def rootMeanSetN (xValue):
        return rootMeanFunc (nValue, xValue)
    
    integral = rm.romberg (waveFuncSetN, (x1, x2), tolerance)
    rootMeanSquared = rootMeanSquaredFunc (errorSamplePoints, nValue, x1, x2)
    
    return integral, rootMeanSquared
```

```python
x1 = -math.pi / 2#-4
x2 = math.pi / 2#4
tolerance = 1e-10
nValues = [0, 1, 2, 3, 4, 5]

for n in nValues:
    integral, error = rombergQuantumWaveFunc (quantumWaveFunc, n, x1, x2, tolerance)
    print ("n:{} Integral: {:f} Root-Mean-Squared: {:f}".format (n, integral, error))
```

These calculations show how the wave function's integral is zero for all even n-values (which is a result of odd versus even functions) and that the error is based upon the energy level of the particle. With this we can see that the error increases with the energy increase.
